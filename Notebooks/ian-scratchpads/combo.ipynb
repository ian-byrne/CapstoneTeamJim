```python
data = pd.read_csv(
    "/Users/ianbyrne/Development/CapstoneTeamJim/notebooks/ian-scratchpads/Data/nn_monthly_data.csv.gz",
    parse_dates=["period_begin", "period_end"],
)
# add time index
data["time_idx"] = data["period_end"].dt.year * 12 + data["period_end"].dt.month
data["time_idx"] -= data["time_idx"].min()
data["month"] = data["period_end"].dt.month.astype(str).astype("category")
```

```python
data.head()
```

```python
#data.info()
```

```python
train19 = data[(data["year"] == 2019) & (data["month"] == "11")]
preds20 = df2[(df2["month"]== 11)]
```

```python
train19 = train19[["county_fips", "region", "state", "property_type", "month", "median_sale_price"]]
```

```python
train19 = train19[train19["property_type"] == "All Residential"]
```

```python
train19["month"] = train19["month"].astype("int")
```

```python
train19.info()
```

```python
preds20.info
```

```python
df = pd.merge(preds20, train19, how="left")
```

```python
df.rename(columns={"median_sale_price": "target19"}, inplace=True)
```

```python
df
```

```python
df["2019to2020increase"] = (df.preds - df.target19)/df.target19
```

```python
df = df.sort_values(by="2019to2020increase", ascending=False)
```
